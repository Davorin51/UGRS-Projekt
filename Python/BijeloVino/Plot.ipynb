```python
import pandas as pd
import matplotlib.pyplot as plt

# Učitavanje .csv datoteke
data = pd.read_csv('TrainingRGB.csv') # zamijenite 'ime_datoteke.csv' s imenom vaše datoteke

# Izrada vremenskog niza za mjerenja
time = range(len(data))

plt.figure(figsize=(10,8))

# Crvena linija
plt.plot(time, data['R'], color="red", label="Crvena")
# Zelena linija
plt.plot(time, data['G'], color="green", label="Zelena")
# Plava linija
plt.plot(time, data['B'], color="blue", label="Plava")
# Ukupna linija
plt.plot(time, data['W'], color="black", label="Ukupno")

plt.legend()
plt.xlabel('Vrijeme')
plt.ylabel('Intenzitet boje')
plt.title('Promjena boje bijelog vina tijekom vremena')
plt.show()
```
